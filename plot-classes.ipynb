```python
import numpy as np
import matplotlib.pyplot as plt

# Define the high-frequency boundary function
def generate_high_frequency_boundary(x1):
    # High-frequency function: mix of polynomial and sinusoidal terms
    return (
        0.5 * x1**3      # Cubic base shape
        - 0.9 * x1**2    # Quadratic component
        + np.sin(3 * x1)  # High-frequency oscillations
        + 0.5 * np.cos(3 * x1)  # Even higher frequency
    )

# Generate x1 values for plotting
x1 = np.linspace(-5, 5, 1000)  # High resolution for smoother plotting
x2 = generate_high_frequency_boundary(x1)  # Calculate boundary

# Plot the boundary with y-axis limits
plt.figure(figsize=(10, 6))
plt.plot(x1, x2, label='High-Frequency Boundary', color='purple')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)  # Add y=0 reference line
plt.axvline(0, color='black', linestyle='--', linewidth=0.8)  # Add x=0 reference line
plt.title('High-Frequency Classification Boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.ylim(-5, 5)  # Limit y-axis to [-5, 5]
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv("/home/robert/projects/cpp/gradual/cmake-build-debug/examples/pre-training.csv")

# Separate data by class
class_0 = data[data["PredictedClass"] == 0]
class_1 = data[data["PredictedClass"] == 1]

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(class_0["Input1"], class_0["Input2"], color="red", label="Class 0", alpha=0.7)
plt.scatter(class_1["Input1"], class_1["Input2"], color="blue", label="Class 1", alpha=0.7)

# Add labels, legend, and title
plt.xlabel("Input1")
plt.ylabel("Input2")
plt.title("Scatter Plot of Inputs Colored by Predicted Class")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
